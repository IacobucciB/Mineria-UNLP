## Coeficiente de Correlacion Lineal
Calcular el coeficiente de correlacion lineal entre los atributos "petallength" y "petalwidth"

### Librerias

```python
import  pandas as pd
import  matplotlib.pyplot as plt
import seaborn as sns
```

### Acceder a los datos

El archivo subido a WebLidi no era accesible. También estaba disponible en la carpeta compartida de Google Drive en formato Excel. Se convirtió a CSV (opción MSDOS) y puede leerse de la siguiente forma.

```python
path = "iris.csv"

data_frame = pd.read_csv(
    path,
    sep=';',
    decimal=",",
)

data_frame.head()
```

```python
print(data_frame.info())
data_frame.describe()
```

```python
correlation_coefficient = data_frame['petallength'].corr(data_frame['petalwidth'])
display(f"El coeficiente de correlación lineal entre 'petallength' y 'petalwidth' es: {correlation_coefficient:.4f}")
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data_frame, x='petallength', y='petalwidth', hue='class', palette='viridis')
plt.title('Scatter Plot of Petal Length vs. Petal Width by Class')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.legend(title='Class')
plt.grid(True)
plt.show()
```
